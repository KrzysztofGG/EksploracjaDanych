```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import STL
from prophet import Prophet
from sklearn.metrics import mean_absolute_percentage_error
```

```python
DATASET_NAME = "CSUSHPISA"
df = pd.read_csv(DATASET_NAME + '.csv')
df['observation_date'] = pd.to_datetime(df['observation_date'])
df.set_index('observation_date', inplace=True)

df.plot(
    xlabel='time',
    ylabel='Index value',
    title='S&P CoreLogic Case-Shiller U.S. National Home Price Index',
)
```

```python
df.reset_index(inplace=True)
df.rename(columns={'observation_date': 'ds', DATASET_NAME: 'y'}, inplace=True)
df.head()
```

```python
train_years = 4
test_years = 1

start_year = df['ds'].dt.year.min()
end_year = df['ds'].dt.year.max()

mape_scores = []

for year in range(start_year, end_year - train_years - test_years + 2):
    train_start = f"{year}-01-01"
    train_end = f"{year + train_years - 1}-12-31"
    test_start = f"{year + train_years}-01-01"
    test_end = f"{year + train_years + test_years - 1}-12-31"

    train_df = df[(df['ds'] >= train_start) & (df['ds'] <= train_end)]
    test_df = df[(df['ds'] >= test_start) & (df['ds'] <= test_end)]

    print(f"\nTraining: {train_start} to {train_end}, Testing: {test_start} to {test_end}")

    model = Prophet()
    model.fit(train_df)


    future = model.make_future_dataframe(periods=12, freq='M')

    forecast = model.predict(future)

    # align forecast and test data
    forecast_test = forecast.tail(len(test_df))  # Get forecast matching test period

    # calculate error
    error = mean_absolute_percentage_error(test_df['y'], forecast_test['yhat'])
    mape_scores.append(error)
    print(f"MAPE: {error:.4f}")

    # plot
    plt.figure(figsize=(10,5))
    plt.plot(train_df['ds'], train_df['y'], label="Train")
    plt.plot(test_df['ds'], test_df['y'], label="Test")
    plt.plot(forecast['ds'], forecast['yhat'], label="Forecast")
    plt.title(f"{train_start} - {train_end} Train | {test_start} - {test_end} Test\nMAPE: {error:.4f}")
    plt.legend()
    plt.grid()
    plt.show()

# summary
print(f"\nAverage MAPE across all windows: {sum(mape_scores)/len(mape_scores):.4f}")
```
